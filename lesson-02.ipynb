#### `while` loop

```python
count = 1

while count <= 5: # condition to keep the loop running
    print(count)
    count += 1 # increment count by 1 in each iteration. count = count + 1
    print(f"Count is now: {count}")
```

#### 2. `range`

```python
for i in range(5): # range(5) generates numbers from 0 to 4
    print(i)
```
