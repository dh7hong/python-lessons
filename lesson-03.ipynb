#### `Lists`

###### 1. Accessing List Elements

```python
fruits = ["apple", "banana", "orange"]

# List indexing
print(f'fruits[0]: {fruits[0]}')
print(f'fruits[1]: {fruits[1]}')
print(f'fruits[2]: {fruits[2]}')
```

```python
fruits = ["apple", "banana", "orange"]

# Negative indexing
print(f'fruits[-1]: {fruits[-1]}')
```

```python
fruits = ["apple", "banana", "orange"]

# List and for loop
print("Using for loop to iterate over fruits:")
for i in range(len(fruits)): # for i in range(3):
    print(f'{fruits[i]} is at index {i}')
    print(f'fruits[{i}]: {fruits[i]}')
    print(f'len(fruits[{i}]): {len(fruits[i])}')
```

```python
# changing list items
fruits = ["apple", "banana", "orange"]

fruits[1] = "grape"

print(f'Changed fruits: {fruits}')
```

```python
# adding items to a list
fruits = ["apple", "banana"]

fruits.append("orange")

print(f'Adding fruits: {fruits}')
```

```python
# removing items
fruits = ["apple", "banana", "orange"]
fruits.remove("banana")

print(f'Removed banana: {fruits}')
```

```python
# popping items
fruits = ["apple", "banana", "orange"]
popped_fruit = fruits.pop(1)
print(f'Popped fruit: {popped_fruit}')
print(f'Fruits after popping: {fruits}')
```

###### List Slicing and Useful List Methods

```python
numbers = [10, 20, 30, 40, 50]

print(f'numbers[1:4]: {numbers[1:4]}')
print(f'numbers[:3]: {numbers[:3]}')
print(f'numbers[2:]: {numbers[2:]}')

print(f'numbers[::-1]: {numbers[::-1]}')

```

```python
# sort
numbers = [50, 20, 40, 10, 30]
numbers.sort()
print(f'Sorted numbers: {numbers}')

numbers.sort(reverse=True)
print(f'Sorted numbers in descending order: {numbers}')
```

```python
# reverse
numbers = [50, 20, 40, 10, 30]
numbers.reverse()
print(f'Reversed numbers: {numbers}')
numbers.reverse()
print(f'Reversed numbers again: {numbers}')
print(numbers)
```

#### Tuples

```python
colors = ("red", "green", "blue") # Tuples are like lists but cannot be changed.
print(f'colors[0]: {colors[0]}') # Accessing tuple items is just like lists.

```

#### Sets

```python
numbers = {1,2,3,4} # Sets store unique items and are unordered.
print(f'numbers: {numbers}')
```

```python
numbers = set([1, 2, 3, 4, 5])
print(f'numbers: {numbers}')
```

```python
numbers = {1,1,1,2,2,3} # Duplicates are removed automatically.

print(numbers)
```

```python
# adding items to set
numbers = {1,2,3}
numbers.add(4)
print(f'Added 4: {numbers}')

# removing items from set
numbers = {1,2,3}
numbers.remove(2)
print(f'Removed 2: {numbers}')
```

#### Dictionaries

```python
# Dictionaries have key-value pairs and are unordered.
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A"
}
```

```python
# Dictionaries have key-value pairs and are unordered.
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A"
}

# Dictionary keys
print(f'student keys: {list(student.keys())}')

# Dictionary values
print(f'student values: {list(student.values())}')
```

```python
# Dictionaries have key-value pairs and are unordered.
student = {
    "name": "Alice",
    "age": 20,
    "grade": "A"
}

# Accessing dictionary items
print(f'student["name"]: {student["name"]}')

print(f'student["age"]: {student["age"]}')

# changing dictionary items
student["age"] = 21
print(f'Changed student age: {student["age"]}')
```

#### Looping through dictionaries

```python
student = {
    "name": "Alice",
    "age": 20
}

for key, value in student.items():
    print(key, value)
```

#### Nested Collections

```python
students = [
    {
        "name": "Alice",
        "grade": "A"
    },
    {
        "name": "Bob",
        "grade": "B"
    }
]

# Chaining indexing to access nested data
print(students[0]["name"])
```
